## Preprocessing

```python
# dropping id column as it is unnecessary.
data.drop('ID_code',axis=1,inplace=True)
```

```python
data
```

### Splitting data

```python
x = data.drop('target',axis=1)
y = data.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

### Handling class imbalance

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_train_smote,y_train_smote = smote.fit_resample(x_train,y_train)
```

```python
from collections import Counter
print(Counter(y_train))
print(Counter(y_train_smote))
```

## Model building

### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(class_weight='balanced')
lr.fit(x_train_smote,y_train_smote)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
x_pred = lr.predict(x_train_smote)
```

```python
print("Logistic regression Training result:")
print(classification_report(x_pred,y_train_smote))
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
print("Logistic Regression Testing result:")
print(classification_report(y_test,y_pred))
```

### 2. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50,max_depth=10,class_weight='balanced')
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred = rfc.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
print("Random Forest Testing Result:")
print(classification_report(y_test,y_pred))
```

### 3. XGBoost

```python
from xgboost import XGBClassifier
```

```python
counter = Counter(y_train)
ratio = counter[0] / counter[1]
```

```python
model = XGBClassifier(scale_pos_weight=ratio,max_depth=6,learning_rate=0.1,n_estimators=300)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print('XGBoost Testing Result')
print(classification_report(y_test,y_pred))
```

### Threshold tuning

```python
probs = model.predict_proba(x_test)[:, 1]   # get probability of class 1
y_pred = (probs > 0.40).astype(int)         # set your own threshold
```

```python
print("XGboost testing result after threshold tuning:")
print(classification_report(y_test,y_pred))
```
